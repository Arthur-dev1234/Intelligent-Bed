```python
import numpy as np
import os
from PIL import Image
import cv2 as cv
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
def read_directory(directory_name):    
    for filename in os.listdir(directory_name):
        img = Image.open(directory_name + "/" + filename)
        arr = image.img_to_array(img)
        array_of_all_img.append(arr)
```

```python
def sep_img_arr_with_cols(img_array_list):
    i = 0
    for arr in img_array_list:
        new_array_after_sep_with_cols.append([])
        for y in range (col):
            new_array_after_sep_with_cols[i].append([])
            for x in range (row):
                new_array_after_sep_with_cols[i][y].append(arr[x][y])
        new_array_after_sep_with_cols[i] = np.array(new_array_after_sep_with_cols[i])
        i += 1
    print(len(new_array_after_sep_with_cols))
```

```python
def pan_operate(new_img_array_list):
    
    for arr in new_img_array_list:    
        col_num = []
        col255_num = []
        for i in range(col):
            if ((arr[i] == 255).all()) == False:
                col_num.append(i)
            else:
                col255_num.append(i)
    
        temp = []
        picnum = 0
        
        for num in range (len(col255_num)+1):
            cols = 0
            temp.append([])
            for x in range(col):
                temp[picnum].append([])
            cols = cols + picnum
            for j in col_num:   
                for n in range(row):
                    temp[picnum][cols].append(arr[j][n])
                cols += 1
            for k in col255_num:
                if cols <= 6:
                    for n in range(row):
                        temp[picnum][cols].append(arr[k][n])
                    cols += 1
                else:
                    cols = 0
                    for n in range(row):
                        temp[picnum][cols].append(arr[k][n])
                    cols += 1
            new_arr = np.array(temp[picnum])
            img_array_after_pan.append(new_arr)
            picnum += 1
    print(len(img_array_after_pan))
```

```python
def return_to_7x9(paned_array_list):
    i = 0
    for arr in paned_array_list:
        new_pan_img_array.append([])
        for y in range (row):
            new_pan_img_array[i].append([])
            for x in range (col):
                new_pan_img_array[i][y].append(arr[x][y])
        new_pan_img_array[i] = np.array(new_pan_img_array[i])
        i += 1
    print(len(new_pan_img_array))
```

```python
array_of_all_img = []

new_array_after_sep_with_cols = []

img_array_after_pan = []

new_pan_img_array = []

print (array_of_all_img, new_array_after_sep_with_cols, img_array_after_pan, new_pan_img_array, sep = '/')
```

```python
org_dir = '../7x9/4/'
save_dir = '../Varieties/Pan/4/'
pose = 'sit'
col = 7
row = 9

print ("0:lie, 1:under, 2:right, 3:left, 4:sit, 5:unknown")
```

```python
read_directory(org_dir)
```

```python
sep_img_arr_with_cols(array_of_all_img)
```

```python
pan_operate(new_array_after_sep_with_cols)
```

```python
return_to_7x9(img_array_after_pan)
```
